```python
# definamos la siguiente función 
def interes_compuesto(principal, rate, years):
    """
    Calcula el interés compuesto con capitalización anual.

    Args:
        principal (float): El monto principal inicial.
        rate (float): La tasa de interés anual en forma decimal (por ejemplo, 0.05 para 5%).
        years (int): El número de años.

    Returns:
        float: El monto total después de aplicar el interés compuesto.

    ---
    
    Calculates compound interest with annual compounding.

    Args:
        principal (float): The initial principal amount.
        rate (float): The annual interest rate as a decimal (e.g., 0.05 for 5%).
        years (int): The number of years.

    Returns:
        float: The total amount after applying compound interest.
    """
    # Fórmula del interés compuesto: A = P * (1 + r)^t
    amount = principal * (1 + rate) ** years
    return amount
    

```
