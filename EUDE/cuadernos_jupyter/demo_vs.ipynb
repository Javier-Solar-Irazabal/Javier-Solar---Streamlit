Importamos unas librerias para este sesion

Las librerías de Python son como cajas de herramientas que ya traen funciones y soluciones listas para usar. En lugar de programar todo desde cero, usamos librerías para ahorrar tiempo y trabajar de forma más sencilla y ordenada. Cada librería está pensada para un tipo de tarea: cálculos, manejo de datos, gráficos, inteligencia artificial, etc.

Dos de las librerías más conocidas y usadas son:

**NumPy**: se usa principalmente para trabajar con números y cálculos matemáticos. Es muy útil cuando necesitamos hacer operaciones con muchos datos al mismo tiempo, como en matemáticas, física o ingeniería.

**Pandas** : se utiliza para organizar, analizar y manipular datos, especialmente en forma de tablas (parecidas a las de Excel). Es muy común en análisis de datos porque permite limpiar información, ordenar datos y sacar conclusiones fácilmente.
